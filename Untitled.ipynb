```python
import pandas as pd
```

```python
books = pd.read_csv('./new_books.csv')
categories = pd.read_csv('./new_categories.csv')
category_books = pd.read_csv('./new_category_books.csv')
```

```python
joined_categories_books = pd.merge(categories, category_books, left_on='id', right_on='category_id')
```

```python
len(joined_categories_books)
```

```python
final_categories_books = pd.merge(books, joined_categories_books, left_on='id', right_on='book_id')
```

```python
len(final_categories_books)
```

```python
final_categories_books.to_csv('joined_categories_books.csv', index=False)
```
